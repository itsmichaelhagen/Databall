```python
def age(birthdate):
    today = dt.date.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age
```

```python
class Player:
    def __init__(self, index, pid, number, first, last, pos, ht, wt, bat, throw, birthdate, organization, team, level):
        self.index = index
        self.pid = pid
        self.number = number
        self.first = first
        self.last = last
        self.pos = pos
        self.ht = ht
        self.wt = wt
        self.bat = bat
        self.throw = throw
        self.birthdate = birthdate
        self.age = (age(self.birthdate))
        self.organization = organization
        self.team = team
        self.level = level
    
```

```python
class battingMLB:
    def __init__(self,g,pa,ab,r,hits,doubles,triples,hr,rbi,bb,so,sb,cs,hbp,tb,gdp,sh,sf,ibb):
        self.g = g
        self.pa = pa
        self.r = r
        self.ab = ab
        self.hits = hits
        self.doubles = doubles
        self.triples = triples
        self.hr = hr
        self.rbi = rbi
        self.bb = bb
        self.so = so
        self.sb = sb
        self.cs = cs
        self.hbp = hbp
        self.tb = tb
        self.gdp = gdp
        self.sh = sh
        self.sf = sf
        self.ibb = ibb
        
        def avgCalc(x,y):
            
            try:
                self.avg = round(x/y,3)
            except ZeroDivisionError:
                self.avg = 0
                
        def obpCalc(a,b,c,d,e):
            
            try:
                self.obp = round((a+c+d)/(b+c+d+e), 3)
            except ZeroDivisionError:
                self.obp = 0 
                
        def slgCalc(x,y):
            
            try:
                self.slg = round(x/y,3)
            except ZeroDivisionError:
                self.slg = 0
                
        def opsCalc():
            try:
                self.ops = round(self.obp+self.slg, 3)
            except ZeroDivisionError:
                self.ops = 0
        
        def sopctCalc():
            try:
                self.sopct = round(self.so/self.ab, 3)
            except ZeroDivisionError:
                self.sopct = 0
            
        avgCalc(hits,ab)
        obpCalc(hits,ab,bb,hbp,sf)
        slgCalc(tb,ab)
        opsCalc()
        sopctCalc()
```

```python
class pitchingMLB:
    def __init__(self,g,gs,w,l,gf,cg,sho,sv,ip,h,r,er,hr,bb,ibb,so,hbp,bk,bf,wp):
        self.g = g
        self.gs = gs
        self.w = w
        self.l = l
        self.gf = gf
        self.cg = cg
        self.sho = sho
        self.sv = sv
        self.ip = ip
        self.h = h
        self.r = r
        self.er = er
        self.hr = hr
        self.bb = bb
        self.ibb = ibb
        self.so = so
        self.hbp = hbp
        self.bk = bk
        self.bf = bf
        self.wp = wp
        
        def eraCalc(x,y):
            try:
                self.era = round((9*x/y), 2)
            except ZeroDivisionError:
                self.era = 0
        
        def wpctCalc(x,y):
            try:
                self.wpct = round((x/(x+y)))
            except ZeroDivisionError:
                self.wpct = 0
        
        def whipCalc(x,y,z):
            try:
                self.whip = round(((x+y)/z),3)
            except ZeroDivisionError:
                self.whip = 0
        
        eraCalc(er,ip)
        wpctCalc(w,l)
        whipCalc(h,bb,ip)
                
```
